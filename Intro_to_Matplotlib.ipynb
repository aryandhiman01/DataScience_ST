```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv")
print(df.head())
```

```python
#Average workout duration
avg = df["Duration"].mean()
print(avg)
```

```python
#Most Common Workout Duration
most_common = df["Duration"].mode()[0]
print(most_common)
```

```python
#Line Chart
plt.figure(figsize=(8,5))
plt.plot(df.index,df["Calories"])
plt.title("Calories Burned per Workout")
plt.xlabel("Workout Number")
plt.ylabel("Calories")
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Duration"], bins=10)
plt.title("Workout Duartion Distribution")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["Duration"], df["Calories"])
plt.title("Duration vs Calories")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.show()
```

```python
#Bar Chart
avg_values = df.mean(numeric_only=True)

plt.figure(figsize=(8,5))
plt.bar(avg_values.index, avg_values.values)
plt.title("Average Values of Variables")
plt.ylabel("Average")
plt.show()
```

```python
#Relationship Between Pulse and Calories
plt.figure(figsize=(8,5))
plt.scatter(df["Pulse"], df["Calories"])
plt.title("Pulse vs Calories")
plt.xlabel("Pulse")
plt.ylabel("Calories")
plt.show()
```
