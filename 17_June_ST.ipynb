```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load Dataset
df = pd.read_csv("data.csv")
df.head()
```

```python
#Simple Scatterplot using default theme
sns.scatterplot(
    data=df,
    x=df.columns[0],
    y=df.columns[1]
)
plt.show()
```

```python
# Available themes
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

for theme in themes:
    sns.set_theme(style=theme)

    plt.figure(figsize=(8, 5))

    sns.scatterplot(
        data=df,
        x="Duration",
        y="Calories",
        hue="Pulse",
        size="Maxpulse",
        palette="viridis"
    )

    plt.title(f"Seaborn Theme: {theme}")
    plt.show()
```

```python
# Color Palettes
palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

for palette in palettes:
    sns.palplot(sns.color_palette(palette))
    plt.title(palette)
    plt.show()
```

```python
# Boxplot
sns.boxplot(
    y=df[df.columns[0]]
)

plt.title("Box Plot")
plt.show()
```

```python
#Violin plot
sns.violinplot(
    y = df[df.columns[0]]
)

plt.title
plt.show()
```

```python
sns.barplot(
    data=df,
    x="Pulse",
    y="Calories",
    hue="Duration"
)
plt.show()
```

```python
#Heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
