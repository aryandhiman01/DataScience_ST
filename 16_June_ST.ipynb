```python
#Ques:21-
plt.figure(figsize=(10,6))
df[["Duration", "Pulse", "Maxpulse", "Calories"]].boxplot()
plt.title("Box Plot of Numerical Variables")
plt.ylabel("Values")
plt.show()

def count_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]

    return len(outliers)
    
for col in ["Duration", "Pulse", "Maxpulse", "Calories"]:
    print(col, ":", count_outliers(col))
```

```python
#Ques:22-
outlier_counts = {}
for col in ["Duration", "Pulse", "Maxpulse", "Calories"]:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    outlier_counts[col] = len(outliers)

print(outlier_counts)
```

```python
#Ques:23-
Q1 = df["Calories"].quantile(0.25)
Q3 = df["Calories"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

#Remove outliers
clean_df = df[
    (df["Calories"] >= lower_bound) &
    (df["Calories"] <= upper_bound)
]

print("Original Shape:", df.shape)
print("Cleaned Shape:", clean_df.shape)

#Graphs
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Original Histogram
plt.subplot(1,2,1)
plt.hist(df["Calories"], bins=15, edgecolor="black")
plt.title("Original Calories Histogram")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(True)

# Cleaned Histogram
plt.subplot(1,2,2)
plt.hist(clean_df["Calories"], bins=15, edgecolor="black")
plt.title("Cleaned Calories Histogram")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```
