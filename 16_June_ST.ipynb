```python
#Matpotlib Exercise Solutions

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv")
print(df.head(11))
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("Columns Name:", df.columns)
print("DataTypes:", df.dtypes)
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Pulse"])
plt.title("Pulse Variable")
plt.xlabel("Workout Record")
plt.ylabel("Pulse Values")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Maxpulse"])
plt.title("Maxpulse Variable")
plt.xlabel("Workout Record")
plt.ylabel("Maxpulse Values")
plt.grid(True)
plt.show()
```

```python
values = df[["Duration", "Pulse", "Maxpulse", "Calories"]].mean()
plt.figure(figsize=(8,5))
values.plot(kind='bar')
plt.title("Average Values of Workout Variables")
plt.xlabel("Variables")
plt.ylabel("Average Values")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Duration"], bins=10)
plt.title("Dration Variable")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Calories"], bins=10)
plt.title("Calories Variable")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Duration"], df["Calories"])
plt.title("Duration and Calories")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Pulse"], label="Pulse")
plt.plot(df["Maxpulse"], label="Maxpulse")
plt.title("Pulse and Maxpulse")
plt.xlabel("Pulse")
plt.ylabel("Maxpulse")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_five = df["Duration"].value_counts().head(5)
plt.figure(figsize=(8,5))
plt.pie(
    top_five,
    labels=top_five.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Top 5 Most Frequent Workout Durations")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["Calories"].dropna())
plt.xlabel("Box plot for the Calories")
plt.ylabel("Calories")
plt.grid(True)
plt.show()
```
