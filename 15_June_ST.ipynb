```python
import pandas as pd
import numpy as np
data={
    'Name':['Arjun','Priya','Rahul','Sneha','Vikram'],
    'Age':[21,22,20,23,21],
    'City':['Delhi','Mumbai','Pune','Bangalore','Bhubaneswar'],
    'Marks':[85,90,78,95,88]
}
df=pd.DataFrame(data)
print(df)
```

```python
df['Percentage']=(df['Marks']/100)*100
print("===After adding percentage column===")
print(df)
```

```python
df.rename(columns={'Marks':'Score','City':'Location'},inplace=True)
print("===After renaming columns===")
print(df)
print(df.columns.tolist())
```

```python
#Drop a Column
```

```python
df_no_pct = df.drop(columns = ['Percentage'])
print("After Dropping Percentage:")
print(df_no_pct)
```

```python
#Drop row by index
df_no_row2 = df.drop(index=2)
print('\nAfter Dropping Row:')
print(df_no_row2)
```

```python
df = pd.DataFrame({
    'Name':['Arjun','Priya','Rahul','Sneha','Vikram'],
    'Score':[85,90,78,95,88],
    'Age':[21,22,20,23,19],
})

#apply() on a single column
df['Grade'] = df['Score'].apply(lambda x: 'A+' if x >= 90 else('A' if x >= 80 else 'B'))
print('\n',df)

#apply o entire row
df['Summary'] = df.apply(lambda row: f"{row['Name']} scored {row['Score']}", axis=1)
print('\n',df['Summary'])

df['Qualification'] = df['Age'].apply(lambda x: 'Graduate' if x >= 21 else('Not Graduate'))
print('\n',df)
```

```python
#map() on a series - replace values using a dictionary
grade_map = {'A+': 'Excellent', 'A': 'Very Good', 'B': 'Good'}
df['Grade_label'] = df['Grade'].map(grade_map)
print(df[['Name', 'Grade', 'Grade_label']])
```
