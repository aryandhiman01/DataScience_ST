```python
# PHASE 1 — DATA CLEANING
```

```python
covid_clean = covid.copy()
```

```python
covid_clean.isnull().sum()
```

```python
missing = pd.DataFrame({

    "Missing Values":
    covid_clean.isnull().sum(),

    "Percentage":
    round(
        covid_clean.isnull().mean()*100,
        2
    )

})

missing
```

```python
covid_clean["Province/State"] = (
    covid_clean["Province/State"]
    .fillna("Unknown")
)
```

```python
covid_clean.isnull().sum()
```

```python
covid_clean.duplicated().sum()
```

```python
covid_clean.duplicated().sum()
```

```python
covid_clean.dtypes
```

```python
covid_clean["Date"] = pd.to_datetime(
    covid_clean["Date"]
)
```

```python
covid_clean["Date"].dtype
```

```python
num_cols = [

    "Lat",
    "Long",
    "Confirmed",
    "Deaths",
    "Recovered",
    "Active"

]

covid_clean[num_cols].describe()
```

```python
covid_clean[
    covid_clean["Confirmed"] < 0
]
```

```python
covid_clean[
    covid_clean["Deaths"] < 0
]
```

```python
covid_clean[
    covid_clean["Recovered"] < 0
]
```

```python
covid_clean[
    covid_clean["Active"] < 0
]
```

```python
covid_clean["Active"] = covid_clean["Active"].clip(lower=0)
```

```python
covid_clean[covid_clean["Active"] < 0]
```

```python
covid_clean["Calculated_Active"] = (

    covid_clean["Confirmed"]

    -

    covid_clean["Deaths"]

    -

    covid_clean["Recovered"]

)
```

```python
covid_clean[
    covid_clean["Calculated_Active"]

    !=

    covid_clean["Active"]
]
```

```python
mismatch_count = (
    covid_clean["Calculated_Active"]
    != covid_clean["Active"]
).sum()

print("Mismatch Rows:", mismatch_count)
```

```python
(covid_clean["Active"] < 0).sum()
```

```python
covid_clean.rename(
    columns={
        "Calculated_Active":
        "Recalculated_Active"
    },
    inplace=True
)
```

```python
covid_clean["Active_Difference"] = (

    covid_clean["Active"]

    -

    covid_clean["Recalculated_Active"]

)
```

```python
covid_clean[
    covid_clean["Active_Difference"] != 0
]
```

```python
(
covid_clean["Active_Difference"] != 0
).sum()
```
